# Hyperparameter Experiments Analysis

This notebook summarizes and analyzes the three experiments conducted on the CIFAR-10 classification task using a Convolutional Neural Network (CNN).

The goal of these experiments was to evaluate the impact of:

- Learning rate
- Batch size

on validation performance.

## Experimental Setup

All experiments were trained under the same conditions:

- Architecture: SimpleCNN
- Optimizer: Adam
- Loss function: CrossEntropyLoss
- Epochs: 5
- Early stopping enabled
- Best model saved based on validation loss

Only the following hyperparameters were modified:

1. Learning rate
2. Batch size

```python
import pandas as pd

data = {
    "Experiment": ["Exp_Condition 1", "Exp_Condition 2", "Exp_Condition 3"],
    "Learning Rate": [0.001, 0.0005, 0.001],
    "Batch Size": [32, 32, 64],
    "Best Epoch": [4, 5, 5],
    "Best Validation Loss": [0.8277, 0.8445, 0.8361],
}

df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt

plt.figure()

plt.plot(df["Experiment"], df["Best Validation Loss"], marker='o')

for i, value in enumerate(df["Best Validation Loss"]):
    plt.text(i, value, f"{value:.4f}", ha='center', va='bottom')

plt.xlabel("Experiment")
plt.ylabel("Best Validation Loss")
plt.title("Validation Loss Comparison Across Experiments")
plt.grid(True)

plt.show()
```

## Learning Rate Analysis

Comparing Experiment 1 (LR = 0.001) and Experiment 2 (LR = 0.0005):

Reducing the learning rate resulted in:

- Slower improvement across epochs
- Slightly higher final validation loss within 5 epochs

This suggests that a lower learning rate may require more training epochs to reach optimal performance.

## Batch Size Analysis

Comparing Experiment 1 (Batch = 32) and Experiment 3 (Batch = 64):

Increasing the batch size resulted in:

- More stable gradient updates
- Slightly fewer parameter updates per epoch
- Slightly higher validation loss compared to batch size 32

With number of epochs (5), a smaller batch size provided more frequent weight updates and slightly better generalization performance.

## Final Conclusion

Among the three experiments, the best performing configuration was:

- Learning Rate: 0.001
- Batch Size: 32
- Best Validation Loss: 0.8277

This configuration provided the best trade-off between convergence speed and generalization within the given training duration.

For having an improvment we should:
- Increasing the number of training epochs
- Testing additional learning rate values.

The file final_best_model.pth is just a copy of the best model from the experiments.
